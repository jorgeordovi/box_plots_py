```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
np.random.seed(0)
data0= [np.random.normal(0, std, 100) for std in range(1, 4)]
```

```python
plt.boxplot(data0)
plt.title('Boxplot Basic')
plt.xlabel('Group')
plt.ylabel('Values')
plt.show
```

```python
plt.violinplot(data0)
plt.title('Boxplot Violin Basic')
plt.xlabel('Group')
plt.ylabel('Values')
plt.show
```

```python
plt.style.use('default')
data2 = [np.random.normal(1, std, 100) for std in range(1, 4)]

fig, axs=plt.subplots(1,2, figsize=(12, 6))
axs[0].boxplot(data0)
axs[0].set_title('Box-plots Group 1')
axs[0].set_xlabel('Group')
axs[0].set_ylabel('Values')


axs[1].boxplot(data2)
axs[1].set_title('Box-plots Group 2')
axs[1].set_xlabel('Group')
axs[1].set_ylabel('Values') 

plt.tight_layout()
plt.show()
```

```python
fig, axs=plt.subplots(1,2, figsize=(12, 6))
axs[0].violinplot(data0)
axs[0].set_title('Boxplot Violin Basic 1')
axs[0].set_xlabel('Group')
axs[0].set_label('Values')

axs[1].violinplot(data2)
axs[1].set_title('Boxplot Violin Basic 2')
axs[1].set_xlabel('Group')
axs[1].set_ylabel('Values')
plt.show
```
